```python
# Import necessary libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
from sklearn import tree
from sklearn.datasets import load_diabetes
```

```python
# Load the Diabetes dataset
diabetes = load_diabetes()
X = diabetes.data  # Features (independent variables)
y = diabetes.target  # Target (dependent variable)
```

```python
# For simplicity, let's binarize the target variable to create a classification problem
# Convert continuous target to binary based on a threshold (e.g., if target > median, classify as 1, else 0)
y_binary = (y > np.median(y)).astype(int)  # Use np.median() instead of y.median()
```

```python
# Step 1: Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_binary, test_size=0.3, random_state=42)
```

```python
# Step 2: Initialize the Decision Tree Classifier
clf = DecisionTreeClassifier(random_state=42, max_depth=5)  # Limit the tree depth to avoid overfitting
```

```python
# Step 3: Train the model on the training set
clf.fit(X_train, y_train)
```

```python
# Step 4: Make predictions on the test set
y_pred = clf.predict(X_test)
```

```python
# Step 5: Evaluate the model (display only accuracy)
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy * 100:.2f}%')
```

```python
# Optional: Visualize the decision tree with a different style
plt.figure(figsize=(12, 8))
tree.plot_tree(clf, filled=True, feature_names=diabetes.feature_names, class_names=['0', '1'],
               fontsize=10, impurity=False, proportion=True)
plt.title("Decision Tree Visualization (Depth Limit: 5)")
plt.show()
```
